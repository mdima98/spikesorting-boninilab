# Probe File Creation

Creation of the probe file needed by SpikeInterface using probeionterface. This may also be inlcude in the pre processing if the new probe fill will be compatible wih kilosort4

```python
import numpy as np
import matplotlib.pyplot as plt

import probeinterface as pi

from probeinterface import Probe, ProbeGroup
from probeinterface.plotting import plot_probe, plot_probe_group
```

```python
square_probe = pi.generate_multi_columns_probe(
    num_columns=12,
    num_contact_per_column=12,
    xpitch=10, ypitch=10,
    contact_shapes='square', contact_shape_params={'width': 8}
)
square_probe.create_auto_shape('rect')
plot_probe(square_probe)

plt.show()
```
